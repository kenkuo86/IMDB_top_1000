```python
import pandas as pd

# import data
imdb = pd.read_csv('imdb_top_1000.csv')
imdb.head()
```

```python
# Data quality check
print(imdb.info())
```

```python
# Convert Runtime into an integer column
runtime = imdb['Runtime'].str.replace('min', '').astype('int')

trans_runtime = imdb.assign( runtime_in_minutes = lambda df_: runtime)
trans_runtime
```

```python
# split movie genres in the Genre column
trans_runtime['Genre'] = trans_runtime['Genre'].str.split(', ')
```

```python
processed = trans_runtime.explode('Genre')
# genre_processed['Series_Title', 'Released_Year', 'Genre', 'IMDB_Rating']
```

```python
# combine all stars into one column

processed['All_Stars'] = processed[['Star1', 'Star2', 'Star3', 'Star4']].values.tolist()
star_processed = processed.explode('All_Stars')
```

```python
star_processed.info()
```

```python
exporting_columns = ['Series_Title', 
                     'Released_Year', 
                     'Certificate', 
                     'Genre', 
                     'All_Stars',
                     'runtime_in_minutes',
                     'IMDB_Rating', 
                     'Meta_score', 
                     'Director',
                     'No_of_Votes',
                     'Gross'
                    ]
export = star_processed[exporting_columns]
export
```

```python
# export data
export.to_csv('IMDB_top_1000_flatten.csv')
```
